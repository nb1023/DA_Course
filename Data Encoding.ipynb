## ONE HOT ENCODING

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
## create ans instance of one hot encoder

encoder = OneHotEncoder()
```

```python
## perform fit and then transform

encoded = encoder.fit_transform(df[['color']]).toarray()
#alphabetical sorting is done the feature swill be - blue, green , red
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## for new data
encoder.transform([['blue']]).toarray()
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['sex','smoker','day','time']]).toarray()
```

```python
encoder_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
pd.concat([df, encoder_df], axis=1)
```

## NOMINAL ENCODING

```python
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df['color'])
```

```python
lbl_encoder.transform([['blue']])
```

## ORDINAL ENCODING

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
df
```

```python
## create an instance of ordinal encoder and then fit_transform

encoder = OrdinalEncoder(categories = [['small', 'medium', 'large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

## TARGET GUIDED ORDINAL ENCODING

```python
df = pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
    'price': [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df
```

```python
df[['price', 'city_encoded']]
```
